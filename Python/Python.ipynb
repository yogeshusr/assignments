### 9. Write a Python function that takes a list and returns a new list with unique elements of the first list.

```python
9. Write a Python function that takes a list and returns a new list with unique elements of the first list.
Sample List : [1,2,3,3,3,3,4,5]
Unique List : [1, 2, 3, 4, 5]
```

```python
def unique_list(l):
  x = []
  for i in l:
    if i not in x:
      x.append(i)
  return x

print(unique_list([1,2,3,3,3,3,4,5]))
```
