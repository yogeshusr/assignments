```python
#Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
#Reading the CSV file
data_book = pd.read_csv('book.csv',encoding='latin-1')
data_book   
```

```python
data_book.head()
```

```python
# Visualizing the frequency for the different book categories
plt.bar(data_book.columns ,data_book.sum(axis=0),edgecolor='green',color='red')
plt.xticks(rotation = 90)
plt.xlabel("Books")
plt.ylabel("Frequency")
plt.show()
```

```python
# Implementing an apriori algorithm
book_apriori = apriori(data_book, min_support=0.1, use_colnames=True)
book_apriori
```

### Constructing an association model, using lift as the evaluation metric

```python
rules = association_rules(book_apriori, metric="lift", min_threshold=0.8)
rules
```

### Sorting : books based on the association rule with respect to lift criteria

```python
rules.sort_values('lift',ascending = False,inplace=True)
rules
```

```python
# Scatter plot for support and confidence for the first association role, with respect to Lift criteria
plt.scatter(rules['support'], rules['confidence'],c=rules['lift'],cmap= 'magma',alpha=0.75)
plt.xlabel('support')
plt.ylabel('confidence')
plt.title('Support vs Confidence')
cbar=plt.colorbar()
cbar.set_label('lift')
plt.show()
```

```python
# Scatter plot for support and lift for the first association role, with respect to Confidence criteria
plt.scatter(rules["support"], rules["lift"],c=rules['confidence'],alpha=0.75)
plt.xlabel("support")
plt.ylabel("lift")
plt.title("Support vs Lift")
cbar=plt.colorbar()
cbar.set_label('confidence')
plt.show()
```

```python
#Checking the Associated books for the category, "ChildBooks"
bookname='ChildBks'
rules_ChildBks =(rules.lift>1) & (rules['antecedents'].astype(str).str.contains(bookname))
rules.loc[rules_ChildBks,["antecedents","consequents"]]
```
